# Отток клиентов

**Ситуация**: из «Бета-Банка» стали уходить клиенты, каждый месяц по-немногу, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

**Задача**: спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. То есть задача *бинарной классификации*. Нам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

**Методы**: построим модель со значением *F1*-меры не менее 0.59. Проверим *F1*-меру на тестовой выборке. Дополнительно измерим *AUC-ROC*, сравнивая её значение с *F1*-мерой.

## Исследование задачи

Для исследования выберем три модели: дерево решений, случайный лес и логистическую регрессию. 

### Дерево решений

```python
f1_max = 0
depth_plt= []
f_plt = []
dep=0
for depth in range(1,30):
    model = DecisionTreeClassifier(random_state=RANDOM_STATE, max_depth=depth)
    model.fit(features_train,target_train)
    valid_predict = model.predict(features_valid)
    result = f1_score(target_valid, valid_predict)
    depth_plt.append(depth)
    f_plt.append(result)
    if result > f1_max:
        f1_max = result
        dep = depth
print('Наибольшая F1-метрика', f1_max, 'при максимальной глубине', dep)
plt.plot(depth_plt,f_plt)
plt.title('Зависимость f1 от глубины дерева')
plt.xlabel('глубина')
plt.ylabel('f1')
plt.grid()
```

Интересно, что начиная примерно с глубины 20 метрика стабилизируется, с увеличением глубины стремится к значению 0.5. 

Построим ROC-кривую и найдем площадь под ней - **AUC-ROC**. Для сравнения на графике дополнительно изображена ROC-кривая для случайной модели - это прямая-диагональ квадрата 1 * 1.

```python
def roc_plot(model, features,target):
    probabilities_valid = model.predict_proba(features_valid)
    probabilities_one_valid = probabilities_valid[:, 1]
    fpr, tpr, thresholds = roc_curve(target_valid,probabilities_one_valid)
    plt.figure()
    plt.plot(fpr,tpr, linestyle = '-')
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.show()
    print('AUC-ROC',roc_auc_score(target_valid, probabilities_one_valid))
```

```python
model = DecisionTreeClassifier(random_state=RANDOM_STATE,
                               max_depth=dep)
model.fit(features_train, target_train)
roc_plot(model, features_valid, target_valid)
```

Итак, F1-метрика для модели *дерева решений* составила 0.582 при максимальной глубине 8, что не удовлетворяет условиям. Площадь под ROC-кривой составила примерно 0.81. 

### Случайный лес

В этом разделе переберем количество деревьев, а также min_samples_leaf для поиска наилучшей модели в смысле метрики F1.

```python
est_plt = []
f_plt = []
leaf_plt = []
best_result = 0
best_est = 0
best_leaf = 0
for est in tqdm(range(3,58)):
    for leaf in range(2,20):
        model = RandomForestClassifier(random_state=RANDOM_STATE, n_estimators=est, min_samples_leaf=leaf)
        model.fit(features_train,target_train)
        
        valid_predict = model.predict(features_valid)
        result = f1_score(target_valid, valid_predict) 
        
        est_plt.append(est)
        f_plt.append(result)
        leaf_plt.append(leaf)
        
        if result > best_result:
            best_model = model
            best_result = result
            best_est = est
            best_leaf = leaf
print('Количество деревьев =', best_est, ', leaf:',best_leaf, ', F1-метрика:', best_result)
```

```python
model = RandomForestClassifier(random_state=RANDOM_STATE, n_estimators=10, min_samples_leaf=3)
model.fit(features_train, target_train)
roc_plot(model, features_valid, target_valid)
```

Итак, F1-метрика для модели случайного леса составила 0.586, что не удовлетворяет условиям. Площадь под ROC-кривой составила примерно 0.84. Это самые высокие результаты из имеющихся, но всё же не достаточно хороши в рамках нашей задачи.

### Логистическая регрессия

```python
model = LogisticRegression(random_state=RANDOM_STATE, solver='liblinear')
model.fit(features_train, target_train)
valid_predict = model.predict(features_valid)
result = f1_score(target_valid, valid_predict) 
print('F1-мера для модели логистической регрессии', result)
roc_plot(model, features_valid, target_valid)
```

Как видим, это самый плохой резульат. Здесь F1-мера составляет всего 0.32, а AUC-ROC 0.79.

**Вывод**: без устранения дисбаланса наибольшей метрикой F1 обладает модель случайного леса с 10 деревьями, хоть и не  удовлетворяет условиям задачи. Чтобы добиться результата, перейдем к устранению дисбаланса классов.

```python
res = pd.DataFrame({'модель': ['дерево решений', 'случайный лес','лог.регрессия'], 'F1':[0.582,0.586,0.321],'AUC-ROC':[0.81,0.837,0.787],'устр.дисбаланса':['нет','нет','нет']})
res
```

## Борьба с дисбалансом

Приступим к борьбе с дисбалансом! Я буду пробовать 4 подхода: взвешивание классов, уменьшение выборки, увеличение выборки, изменение порога.

### Взвешивание классов

Придадим объектам редкого класса больший вес через аргумент class_weight.

#### Дерево решений

```python
f1_max = 0
depth_plt= []
f_plt = []
dep=0
for depth in range(1,30):
    model = DecisionTreeClassifier(random_state=RANDOM_STATE, max_depth=depth, class_weight = 'balanced')
    model.fit(features_train,target_train)
    
    valid_predict = model.predict(features_valid)
    result = f1_score(target_valid, valid_predict)
    
    depth_plt.append(depth)
    f_plt.append(result)
    
    if result > f1_max:
        f1_max = result
        dep = depth
print('Наибольшая F1-метрика', f1_max, 'при максимальной глубине', dep)
plt.plot(depth_plt,f_plt)
plt.title('Зависимость f1 от глубины дерева')
plt.xlabel('глубина')
plt.ylabel('f1')
plt.grid()
```

Возьмем модель с глубиной, доставляющей наиболее значение F1, и построим для неё ROC-кривую.

```python
model = DecisionTreeClassifier(random_state=RANDOM_STATE,
                               max_depth=dep)
model.fit(features_train, target_train)
roc_plot(model, features_valid, target_valid)
```

Здесь значение F1 меньше, чем без взвешивания классов, однако площадь под графиком ROC-кривой больше. Но значения всё еще не соответствуют требованиям.

#### Случайный лес

```python
est_plt = []
f_plt = []
leaf_plt = []
best_result = 0
best_est = 0
best_leaf = 0
for est in tqdm(range(3,58)):
    for leaf in range(2,20):
        model = RandomForestClassifier(random_state=RANDOM_STATE, n_estimators=est, min_samples_leaf=leaf,class_weight = 'balanced')
        model.fit(features_train,target_train)
        valid_predict = model.predict(features_valid)
        
        result = f1_score(target_valid, valid_predict) 
        
        est_plt.append(est)
        f_plt.append(result)
        leaf_plt.append(leaf)
        
        if result > best_result:
            best_model = model
            best_result = result
            best_est = est
            best_leaf = leaf
print('Количество деревьев =', best_est, ', leaf:',best_leaf, ', F1-метрика:', best_result)
```

```python
model = RandomForestClassifier(random_state=RANDOM_STATE, n_estimators=48, min_samples_leaf=7)
model.fit(features_train, target_train)
roc_plot(model, features_valid, target_valid)
```

### Увеличение выборки

Определим функцию *upsample*, которая будет увеличивать выборку. Ранее мы установили, что количество ушедших клиентов в  4 раза меньше оставшихся, поэтому в аргумент функции `repeat` будем передавать 4, чтобы количество оставшихся/ушедших было примерно равным, таким образом добьемся баланса классов.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)#расширенные признаки
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=12345)#перемешиваем
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```

#### Дерево решений

```python
f1_max = 0
depth_plt= []
f_plt = []
dep=0
for depth in range(1,30):
    model = DecisionTreeClassifier(random_state=RANDOM_STATE, max_depth=depth)
    model.fit(features_upsampled,target_upsampled)
    
    valid_predict = model.predict(features_valid)
    result = f1_score(target_valid, valid_predict)
    
    depth_plt.append(depth)
    f_plt.append(result)
    
    if result > f1_max:
        f1_max = result
        dep = depth
print('Наибольшая F1-метрика', f1_max, 'при максимальной глубине', dep)
plt.plot(depth_plt,f_plt)
plt.title('Зависимость f1 от глубины дерева')
plt.xlabel('глубина')
plt.ylabel('f1')
plt.grid()
```

### Уменьшение выборки

Аналогично увеличению выборки, определим функцию *downsample*, которая убирает отрицательные объекты в целевом признаке случайным образом. Аргумент fraction возьмем 0.25, поскольку количество ушедших клиентов в 4 раза меньше, чем оставшихся. Поэтому и количество "нулей" надо уменьшать в 4 раза.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=RANDOM_STATE)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=RANDOM_STATE)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=RANDOM_STATE)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)
```

#### Дерево решений

```python
f1_max = 0
depth_plt= []
f_plt = []
dep=0
for depth in range(1,30):
    model = DecisionTreeClassifier(random_state=RANDOM_STATE, max_depth=depth)
    model.fit(features_downsampled, target_downsampled)
    
    valid_predict = model.predict(features_valid)
    result = f1_score(target_valid, valid_predict)
    
    depth_plt.append(depth)
    f_plt.append(result)
    
    if result > f1_max:
        f1_max = result
        dep = depth
print('Наибольшая F1-метрика', f1_max, 'при максимальной глубине', dep)
plt.plot(depth_plt,f_plt)
plt.title('Зависимость f1 от глубины дерева')
plt.xlabel('глубина')
plt.ylabel('f1')
plt.grid()
```

```python
model = DecisionTreeClassifier(random_state=RANDOM_STATE,
                               max_depth=dep)
model.fit(features_downsampled, target_downsampled)
roc_plot(model, features_valid, target_valid)
```

Ура! У дерева решений наконец достигнут нужный уровень F1-метрики, примерно 0.593.

#### Случайный лес
